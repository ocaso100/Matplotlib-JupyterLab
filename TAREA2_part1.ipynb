```python
import numpy
import cv2
import matplotlib.pyplot as plt  
plt.rcParams['figure.figsize'] = [10,5] #sirve para modificar el tamaño del grafico
```

```python
def linspace(to,tf,N):
    razon=(tf-to)/N
    ls=[]
    for val in range(N):
        ls.append(to+val*razon)
    return ls
```

```python
"""
ECUACION DE LA FUNCION
"""
def cuadratica(g,xo):
    y=[]
    for x in g:   
        y.append((x-xo)**2)
        
    return y
```

```python
"""
GRAFICA DE LA FUNCION
"""
xo=int(input("Ingrese el desplazamiento:"))
a=int(input("Ingrese tf"))
b=int(input("Ingrese to:"))
c=int(input("Ingrese N:"))

t=linspace(a,b,c)#-10,12,15
y=cuadratica(t,xo)
plt.plot(t,y)
plt.title("FUNCIÓN CUADRATICA")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
plt.show()
```
