```python
import numpy
import cv2
import matplotlib.pyplot as plt 
```

```python
def linspace(to,tf,N):
    razon=(tf-to)/N
    ls=[]
    for val in range(N):
        ls.append(to+val*razon)
    return ls
```

```python
"""
ECUACION DE LA FUNCION
"""
def funcion(g):
    y=[]
    for x in g:
        if x>=0: 
            y.append(1/(1+(2.71**(-x)))) 
        if -2<=x and x<0: 
            y.append(0.1*x)
        if x<-2: 
            y.append(0)    
    return y
```

```python
t=linspace(-4,4,10000)#-10,12,15
y=funcion(t)
plt.plot(t,y)
plt.title("FUNCIÓN CUADRATICA")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
plt.show()
```
