```python
import numpy
import cv2
import matplotlib.pyplot as plt 
```

```python
def linspace(to,tf,N):
    razon=(tf-to)/N
    ls=[]
    for val in range(N):
        ls.append(to+val*razon)
    return ls

def desviacion_estandar(valores,media):
    suma=0
    
    for valor in valores:
        suma += (valor-media)**2
        
    radicando= suma/(len(valores)-1)
    
    return radicando**(1/2)

def calcular_media(valores):
    suma=0
    
    for valor in valores: 
        suma += valor
    
    return suma/len(valores)

def funcionP(g,u,ùúé):
    y=[]
    for x in g:
        
        #PUSE EL RAIZ DE 2PI como = 6.28 
        y.append((2.71**(-((x-u)**2)/(2*(ùúé**2))))/(6.28*ùúé))
    return y
                              
```

```python
x=linspace(-10,20,100)

media = calcular_media(x)
print("La media es:",media)
desviacion=desviacion_estandar(x,media)
print("La desviacion estandar es:",desviacion)
```

```python
y=funcionP(x,media,desviacion)
plt.plot(x,y)
plt.title("GAUSS PROBLEMA N¬∞3 ")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
plt.show()
```
